## Paginando os repositórios

```python
api_base_url = 'https://api.github.com'
owner = 'amzn' # username de quem vamos extrair os dados
url = f'{api_base_url}/users/{owner}/repos'

url
```

```python
repos_list = []
for page_num in range(1, 6):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

## Armazenando os dados

### Criando repositório como POST

```python
api_base_url = 'https://api.github.com'
url = f'{api_base_url}/user/repos'

url
```

```python
# Dados para a criação do repositório
data = {
    'name' : 'linguagens-utilizadas',
    'description': 'Repositorio com as linguagens de prog da Amazon',
    'private': False
}

# Enviando a requisição POST para criar o repositório
response = requests.post(url, json=data, headers=headers)

# Verificando o status da resposta
if response.status_code == 201:
    print("Repositório criado com sucesso!")
elif response.status_code == 401:
    print("Erro de autenticação: 401 Unauthorized. Verifique o token de acesso.")
    print(response.json()) # Imprime a resposta JSON para entender melhor o erro
else:
    print(f"Erro ao criar repositório: {response.status_code}")
    print(response.json()) # Imprime a resposta JSON para entender melhor o erro
```

## Upload de arquivo com PUT

```python
api_base_url = 'https://api.github.com'
username = 'JeffAirData'
repo = 'linguagens-utilizadas'
path = 'amazon.csv'

url = f'{api_base_url}/repos/{username}/{repo}/contents/{path}'
url
```

```python
data = {
    'message': 'Adicionando um novo arquivo',
    'content': encoded_content.decode('utf-8')
}

response = requests.put(url, json=data, headers=headers)
# Verificando o status da resposta
if response.status_code == 201:
    print("Repositório criado com sucesso!")
elif response.status_code == 401:
    print("Erro de autenticação: 401 Unauthorized. Verifique o token de acesso.")
    print(response.json()) # Imprime a resposta JSON para entender melhor o erro
else:
    print(f"Erro ao criar repositório: {response.status_code}")
    print(response.json()) # Imprime a resposta JSON para entender melhor o erro

response.status_code
```

```python
# Primeiro, obtenha o SHA do arquivo existente
response = requests.get(url, headers=headers)

if response.status_code == 200:
    file_info = response.json()
    sha = file_info['sha']

    # Preparando os dados para atualizar o arquivo
    data = {
        'message': 'Atualizando o arquivo existente',
        'content': encoded_content.decode('utf-8'),
        'sha': sha  # Incluindo o SHA para atualizar o arquivo existente
    }

    # Fazendo a requisição PUT para atualizar o arquivo
    response = requests.put(url, json=data, headers=headers)

    # Verificando o status da resposta
    if response.status_code == 200:
        print("Arquivo atualizado com sucesso!")
    elif response.status_code == 201:
        print("Arquivo criado com sucesso!")
    elif response.status_code == 401:
        print("Erro de autenticação: 401 Unauthorized. Verifique o token de acesso.")
        print(response.json())  # Imprime a resposta JSON para entender melhor o erro
    else:
        print(f"Erro ao atualizar o arquivo: {response.status_code}")
        print(response.json())  # Imprime a resposta JSON para entender melhor o erro

else:
    print(f"Erro ao obter informações do arquivo: {response.status_code}")
    print(response.json())  # Imprime a resposta JSON para entender melhor o erro
```
